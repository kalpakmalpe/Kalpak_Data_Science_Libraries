```python
import pandas as pd
df1=pd.DataFrame({'roll no':[1,2,3,4,5],'name':['a1','a2','a3','a4','a5'],'Subject':['sub1','sub2','sub4','sub6','sub5']})
print(df1)
print("\n")

df2=pd.DataFrame({'roll no':[1,2,3,4,5],'name':['b1','b2','b3','b4','b5'],'Subject':['sub2','sub4','sub3','sub6','sub5']})
print(df2)
print("\n")
df3=pd.merge(df1,df2)
print("merge df1 and df2\n",df3)
print("\n")
df3=pd.merge(df1,df2,on='Subject',how='inner')
print("Display sam subject:\n",df3)
print("\n")
df3=pd.merge(df1,df2,on='Subject',how='outer')
print("Display all subject:\n",df3)
print("\n")
df3=pd.merge(df1,df2,on='Subject',how='left')
print("\n")
df4=pd.merge(df1,df2,on='Subject',how='right')
print(df4)
print("\n")


df5=pd.merge(df1,df2,on='Subject',how='right',sort='True')
print(df5)
#df3=pd.merge(df1,df2,on='Subject',how='left')
#print(df3)

#Concat operations
a=pd.concat([df1,df2])
print(a)
```

```python
# Python program to concatenate 
# dataframes using Panda 
  
# Creating first dataframe 
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'], 
                    'D': ['D0', 'D1', 'D2', 'D3']}, 
                    index = [0, 1, 2, 3]) 
  
# Creating second dataframe 
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B4', 'B5', 'B6', 'B7'], 
                    'C': ['C4', 'C5', 'C6', 'C7'], 
                    'D': ['D4', 'D5', 'D6', 'D7']}, 
                    index = [4, 5, 6, 7]) 
  
# Creating third dataframe 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'], 
                    'B': ['B8', 'B9', 'B10', 'B11'], 
                    'C': ['C8', 'C9', 'C10', 'C11'], 
                    'D': ['D8', 'D9', 'D10', 'D11']}, 
                    index = [8, 9, 10, 11]) 
  
# Concatenating the dataframes 
pd.concat([df1, df2, df3]) 
```

```python
# Concatenating the dataframes 
pd.concat([df1, df2, df3]) 

# Python program to merge 
# dataframes using Panda 
  
# Dataframe created

left = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 
                    'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3']}) 
  
right = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'], 
                      'C': ['C0', 'C1', 'C2', 'C3'], 
                        'D': ['D0', 'D1', 'D2', 'D3']}) 
                        
# Merging the dataframes                       
pd.merge(left, right, how ='inner', on ='Key')

# Joining the dataframes                       
left.join(right) 

```
