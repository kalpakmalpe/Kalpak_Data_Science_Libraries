```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(8,10),facecolor=(0,1,0,0.3))
ax1=fig.add_subplot(2,2,1)
ax1.set(xlim=[1,5],ylim=[1,5],xlabel='x-axis',ylabel='y-axis',title='matplotlib')

ax2=fig.add_subplot(2,2,2)
ax2.set(xlim=[1,5])
ax2.set(ylim=[1,5])
ax2.set(xlabel='x-axis')
ax2.set(ylabel='y-axis')
ax2.set(title='matplotlib1')

ax3=fig.add_subplot(2,2,3)
ax3.set(xlim=[1,5],ylim=[1,5],xlabel='x-axis',ylabel='y-axis',title='matplotlib2')

ax4=fig.add_subplot(2,2,4)
ax4.set(xlim=[1,5],ylim=[1,5],xlabel='x-axis',ylabel='y-axis',title='matplotlib3')
plt.show()
```
